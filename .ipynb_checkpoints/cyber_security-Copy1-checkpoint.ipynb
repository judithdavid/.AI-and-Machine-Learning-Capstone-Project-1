```python
#dropping 'double_slash_redirecting' 
df = df.drop('double_slash_redirecting', axis=1)
```

```python
# Building Classification Model
```

```python
#1.Finally, build a robust classification system that classifies whether the URL sample is a phishing site or not.
```

```python
# Build classification models using a binary classifier to detect malicious or phishing URLs.
```

```python
#Identify x and y
```

```python
features = ['index', 'having_IPhaving_IP_Address', 'URLURL_Length',
       'Shortining_Service', 'having_At_Symbol',
       'Prefix_Suffix', 'having_Sub_Domain', 'SSLfinal_State',
       'Domain_registeration_length', 'Favicon', 'port', 'HTTPS_token',
       'Request_URL', 'URL_of_Anchor', 'Links_in_tags', 'SFH',
       'Submitting_to_email', 'Abnormal_URL', 'Redirect', 'on_mouseover',
       'RightClick', 'popUpWidnow', 'Iframe', 'age_of_domain', 'DNSRecord',
       'web_traffic', 'Page_Rank', 'Google_Index', 'Links_pointing_to_page',
       'Statistical_report']
target = ['Result']
```

```python
x = df[features]
y = df[target]
```

```python
#splitting 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
import statsmodels.api as sm
```

```python
# c= sm.add_constant(X_train)
# model = sm.Logit(y_train, X_train).fit()
# model.summary()
# df1 = df.drop('index', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression    
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
pred=logreg.predict(X_test)
```

```python
logreg.score(X_train,y_train)
```

```python
logreg.score(X_test,y_test)
```

```python
#testing
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

```python
# Illustrate the diagnostic ability of this binary classifier by plotting the ROC curve.
```

```python
from sklearn import metrics

y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

```

```python
# Validate the accuracy of data by the K-Fold cross-validation technique.
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import absolute
from numpy import sqrt
```

```python
#define cross-validation method to use
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
#use k-fold CV to evaluate model
scores = cross_val_score(logreg, x, y, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)
```

```python
#view mean absolute error
mean(absolute(scores))
```

```python
# average absolute error between the model prediction and the actual observed data is 0.154
```

```python
#view RMSE
sqrt(mean(absolute(scores)))
```

```python
# root mean squared error (RMSE) was 0.392
```
